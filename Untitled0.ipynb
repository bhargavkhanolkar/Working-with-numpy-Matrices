```python
import numpy as np

# Load the data into NumPy matrices
male = np.genfromtxt('nhanes_adult_male_bmx_2020.csv', delimiter=',', skip_header=1)
female = np.genfromtxt('nhanes_adult_female_bmx_2020.csv', delimiter=',', skip_header=1)

# Check the shape of the matrices
print("Male Matrix Shape: ", male.shape)
print("Female Matrix Shape: ", female.shape)

```

```python
import matplotlib.pyplot as plt

female_weights = female[:, 0]  # First column is weight for females
male_weights = male[:, 0]      # First column is weight for males

# Create subplots for histograms
fig, axes = plt.subplots(2, 1, figsize=(8, 8))

# Top subplot: Female weights
axes[0].hist(female_weights, bins=30, color='purple', alpha=0.7)
axes[0].set_title("Female Weights")
axes[0].set_xlim([40, 150])  # Setting x-axis limits

# Bottom subplot: Male weights
axes[1].hist(male_weights, bins=30, color='blue', alpha=0.7)
axes[1].set_title("Male Weights")
axes[1].set_xlim([40, 150])  # Same x-axis limits

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([female_weights, male_weights], labels=["Female", "Male"], vert=False)
plt.title("Boxplot of Weights for Female and Male Participants")
plt.xlabel("Weight (kg)")
plt.show()
```

```python
import scipy.stats as stats

# Compute basic aggregates for male and female weights
female_mean = np.mean(female_weights)
male_mean = np.mean(male_weights)
female_std = np.std(female_weights)
male_std = np.std(male_weights)
female_skew = stats.skew(female_weights)
male_skew = stats.skew(male_weights)

print(f"Female Mean Weight: {female_mean}, Male Mean Weight: {male_mean}")
print(f"Female Std Weight: {female_std}, Male Std Weight: {male_std}")
print(f"Female Skewness: {female_skew}, Male Skewness: {male_skew}")
```

```python
female_height = female[:, 1]  # Second column is height for females

# Calculate BMI and add as a new column
female_bmi = female_weights / (female_height / 100) ** 2
female = np.column_stack((female, female_bmi))

# Check the updated female matrix
print(female[:5, :])  # Print first 5 rows for verification
```

```python
# Standardize the female matrix (z-scores)
zfemale = (female - np.mean(female, axis=0)) / np.std(female, axis=0)

# Check the standardized dataset
print(zfemale[:5, :])  # Print first 5 rows for verification
```

```python
import seaborn as sns
import pandas as pd

# Create a DataFrame for the standardized data
df_zfemale = pd.DataFrame(zfemale, columns=["Weight", "Height", "Upper Arm Length", "Upper Leg Length",
                                            "Arm Circumference", "Hip Circumference", "Waist Circumference", "BMI"])

# Pairplot for selected variables
sns.pairplot(df_zfemale[["Height", "Weight", "Waist Circumference", "Hip Circumference", "BMI"]])
plt.show()

# Compute Pearson's and Spearman's correlation coefficients
pearson_corr = df_zfemale.corr(method='pearson')
spearman_corr = df_zfemale.corr(method='spearman')

print("Pearson Correlation Coefficients:\n", pearson_corr)
print("Spearman Correlation Coefficients:\n", spearman_corr)
```

```python
# Compute waist-to-height ratio and waist-to-hip ratio
waist_to_height_female = female[:, 6] / female[:, 1]  # Waist circumference to height
waist_to_hip_female = female[:, 6] / female[:, 5]     # Waist circumference to hip circumference

waist_to_height_male = male[:, 6] / male[:, 1]  # Waist circumference to height
waist_to_hip_male = male[:, 6] / male[:, 5]     # Waist circumference to hip circumference

# Add the new columns to the matrices
female = np.column_stack((female, waist_to_height_female, waist_to_hip_female))
male = np.column_stack((male, waist_to_height_male, waist_to_hip_male))

# Check the updated female and male matrices
print(female[:5, :])
print(male[:5, :])
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([waist_to_height_female, waist_to_hip_female, waist_to_height_male, waist_to_hip_male],
            labels=["Female Waist/Height", "Female Waist/Hip", "Male Waist/Height", "Male Waist/Hip"])
plt.title("Boxplot of Waist-to-Height and Waist-to-Hip Ratios")
plt.ylabel("Ratio")
plt.show()
```

```python
# Sort the indices of female participants by BMI
sorted_indices = np.argsort(female_bmi)

# Get the 5 lowest and 5 highest BMI participants
lowest_bmi_female = zfemale[sorted_indices[:5]]
highest_bmi_female = zfemale[sorted_indices[-5:]]

# Print the results
print("Standardized Measurements for 5 Participants with Lowest BMI:")
print(lowest_bmi_female)

print("\nStandardized Measurements for 5 Participants with Highest BMI:")
print(highest_bmi_female)
```
